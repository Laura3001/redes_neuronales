```python
import tensorflow as tf
import os
# import skimage.data as imd (opcion que causa problemas)
import skimage.io as imd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def load_ml_data(data_directory):
        dirs = [d for d in os.listdir(data_directory)
            # listar los directorios
            if os.path.isdir(os.path.join(data_directory, d))]
        # recorrer archivos dentro del directorio
        print(dirs)
        labels = []
        images = []
        for d in dirs:
            label_dir = os.path.join(data_directory, d)
            # Guardar el nombre del archivo
            files_names = [os.path.join(label_dir, f)
                      for f in os.listdir(label_dir)
                      if f.endswith('.ppm')]
            print(label_dir)
            print(files_names)
            
            # Guardar la imagen
            for f in files_names:
                images.append(imd.imread(f))
                labels.append(int(d))
        return images, labels
```

```python
from skimage import transform
w = 9999
h = 9999
for image in images:
    if image.shape[0] < h:
        h = image.shape[0]
    if image.shape[1] < w:
        w = image.shape[1]

print("Tamaño mínimo: {0} x {1}".format(h, w))
```

```python
images30 = [transform.resize(image,(30,30)) for image in images]
```

```python
images30[0]
```

```python
from skimage.color import rgb2gray
images30 = np.array(images30)
images30 = rgb2gray(images30)
```

### Paso 1: Definir la red neuronal

```python
# Valores de entrada a la red neuronal
x = tf.placeholder(dtype = tf.float32, shape = [None, 30, 30])

# Parámetros de salida
y = tf.placeholder(dtype = tf.int32, shape = [None])

# Capa de entrada la red neuronal
images_flat = tf.contrib.layers.flatten(x)

# Capa de conexión
logits = tf.contrib.layers.fully_connected(images_flat, 62, tf.nn.relu)

# Función de pérdidas
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels = y, logits = logits))

# Función de optimización
train_opt = tf.train.AdamOptimizer(learning_rate = 0.001).minimize(loss)

# Convertir el resultado de la red neuronal a la predicción
final_pred = tf.argmax(logits, 1)

# Ajustar la exactitud del modelo de la red neuronal
accuracy = tf.reduce_mean(tf.cast(final_pred, tf.float32))
```

### Paso 2: Entrenamiento de la red neuronal

```python
tf.set_random_seed(1234)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

# Ejecutar la red neuronal con un entrenamiento de 500 iteraciones
# Entrenamos el valor neuronal en base al nivel de EFICACIA (accuracy)
for i in range(501):
    _, accuracy_val = sess.run( [train_opt, accuracy], feed_dict = {
                              x: images30,
                              y: list(labels) }
                              )
    if i%50 == 0:
        print("Eficacia: ", accuracy_val)
```

### Paso 3: Evaluación de la red neuronal

```python
sample_idx = random.sample(range(len(images30)), 10 ) # Se pueden modificar los valores de la cantidad de imágenes a analizar
sample_images = [images30[i] for i in sample_idx]
sample_labels = [labels[i] for i in sample_idx]
```

```python
prediction = sess.run([final_pred], feed_dict = {x:sample_images})[0]
```
